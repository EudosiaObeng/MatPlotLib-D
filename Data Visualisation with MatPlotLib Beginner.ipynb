Scatter plot activity
Create a scatter plot with matplotlib that displays xt paired with yt and zt:

```python
xt=np.array([0,0.1,0.2,0.3,0.5,0.8,1.0])
yt=np.array([1.0,2.1,3.5,6.5,7.2,5.9,6.3])
zt= xt*yt

plt.scatter(xt,yt,marker="s",color="yellow",label="s")
plt.scatter(xt,zt,marker="^",color="blue",label="^")
plt.title("Scatter Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.legend()
plt.grid()
plt.show()

```

```python
#data
countries=['Ghana','Portugal','South Korea','Uruguay','Germany']
gold=[46,27,26,19,17]
silver=[37,23,18,18,10]
bronze=[38,30,25,18,24]
```

```python
plt.figure(figsize = (5,5))
plt.title("Trophy number and type won by 5 countries in Olympics 2016")

countries=["Ghana","Portugal","South Korea","Uruguay","Germany"]
gold=[46,27,26,19,17]
silver=[37,23,18,18,10]
bronze=[38,30,25,18,24]

width=0.2
x=np.arange(len(countries))

plt.bar(x-0.2,gold,width,color="red",label="Gold")
plt.bar(x,silver,width,color="black",label="Silver")
plt.bar(x+0.2,bronze,width,color="orange",label="Bronze")

plt.ylabel("Number of trophies won")
plt.xlabel("Participating Countries")
plt.xticks(x,["Ghana","Portugal","South Korea","Uruguay","Germany"])
plt.legend()

plt.show()
```

Insights made
1.Ghana won the most trophies among the 5 countries.
2.Germany won the least trophies.
